```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi

# Initialize the Kaggle API
api = KaggleApi()
api.authenticate()

# Specify the dataset you want to download
dataset = "taweilo/loan-approval-classification-data"

# Specify the download path
download_path = "/Users/michaelgyimadu/machine-learning/ml-playground/datasets/loan-approval-classification-data"

# Create the directory if it doesn't exist
os.makedirs(download_path, exist_ok=True)

# Download and unzip the dataset
api.dataset_download_files(dataset, path=download_path, unzip=True)

print(f"Dataset downloaded and extracted to: {os.path.abspath(download_path)}")
```

```python
import pandas as pd
import os

def load_data(DOWNLOAD_PATH=download_path):
    csv_path = os.path.join(download_path, "loan_data.csv")
    return pd.read_csv(csv_path)

loan_data = load_data()
loan_data.head()
```
