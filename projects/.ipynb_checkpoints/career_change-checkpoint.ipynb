```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi

# Initialize the Kaggle API
api = KaggleApi()
api.authenticate()

# Specify the dataset you want to download
dataset = "jahnavipaliwal/field-of-study-vs-occupation"

# Specify the download path
download_path = "/Users/michaelgyimadu/machine-learning/ml-playground/datasets/field_of_study_vs_occupation"

# Create the directory if it doesn't exist
os.makedirs(download_path, exist_ok=True)

# Download and unzip the dataset
api.dataset_download_files(dataset, path=download_path, unzip=True)

print(f"Dataset downloaded and extracted to: {os.path.abspath(download_path)}")
```

```python
import pandas as pd
import os

career_change_dataset = os.path.join(download_path, "career_change_prediction_dataset.csv")
df = pd.read_csv(career_change_dataset)

df.head()
```

```python
print(f"Shape: {df.shape}")

df.info()
df.describe()
```

```python
print(df.isnull().sum())
```

```python
print("Column names:", df.columns.tolist())
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df.hist(bins=50, figsize=(20,15))
plt.tight_layout()
plt.show()
```

```python
occupation_counts = df["Current Occupation"].value_counts()
print("Occupation distribution:")
print(occupation_counts)
```

```python
for col in df.select_dtypes(include=['object']).columns:
    print(f"Unique values in {col}:")
    print(df[col].value_counts())
    print()
```

```python
import seaborn as sns

df_numeric = df.select_dtypes(include=['float64', 'int64'])
corr = df_numeric.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
